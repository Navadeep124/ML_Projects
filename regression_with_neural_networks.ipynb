```python
5555555]import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import the dataset
insurance = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
insurance.head()
```

```python
insurance_one_hot = pd.get_dummies(insurance).astype(int)
insurance_one_hot.head()
```

```python
insurance_one_hot.info()
```

```python
x = insurance_one_hot.drop("charges", axis = 1)
y = insurance_one_hot["charges"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
tf.random.set_seed(42)

#Build the model
insurance_model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

#compile the model
insurance_model_1.compile(loss = tf.keras.losses.mae,
                          optimizer=tf.keras.optimizers.SGD(),
                          metrics=['mae'])

#fit the model
insurance_model_1.fit(x_train,y_train,epochs=100)
```

```python
#checking the results of the insurance model
insurance_model_1.evaluate(x_test,y_test)
```

```python
#change some parameters of the model
tf.random.set_seed(42)

insurance_model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

#compile the mmodel
insurance_model_2.compile(loss=tf.keras.losses.mae,
                          optimizer=tf.keras.optimizers.Adam(),
                          metrics=["mae"])

#fit the model
insurance_model_2.fit(x_train, y_train, epochs=200)
```

```python
#evaluate the model
insurance_model_2.evaluate(x_test, y_test)
```

```python
#normalising
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder

#create column transformer
ct = make_column_transformer(
    (MinMaxScaler(), ["age", "bmi", "children"]),
    (OneHotEncoder(handle_unknown="ignore"), ["sex","smoker", "region"])
)

x = insurance.drop("charges", axis=1)
y = insurance["charges"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#fit column tranformer
ct.fit(x_train)

#Transform training and test with normalisation and one hot encoding
x_train_normal = ct.transform(x_train)
x_test_normal = ct.transform(x_test)
```

```python
insurance_model_3 = insurance_model_2

insurance_model_3.fit(x_train_normal, y_train, epochs=200,verbose=0)
insurance_model_3.evaluate(x_test_normal, y_test)
```
